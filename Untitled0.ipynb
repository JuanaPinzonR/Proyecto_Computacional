# **Identificación de Datos Atípicos**

**Método del Rango Intercuartil (IQR)**

se basa en la diferencia entre el primer cuartil (Q1) y el tercer cuartil (Q3) de los datos. Se considera que un valor es atípico si está fuera del rango definido por:

image.png

En la función detectar_iqr, se calcularon Q1, Q3 e IQR. Luego, se filtraron los valores que estuvieran por debajo de
Q1−1.5×IQR o por encima de Q3+1.5×IQR:

```python

def detectar_iqr(df, columna):
    """Método del rango intercuartil (IQR)"""
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[columna] < (Q1 - 1.5 * IQR)) | (df[columna] > (Q3 + 1.5 * IQR))]
    return outliers
```

**Método de la Desviación Estándar**

Este método se basa en la distribución normal de los datos. Se considera que un valor es atípico si se encuentra a más de tres desviaciones estándar por encima o por debajo de la media:

image.png

En la función detectar_desviacion_estandar, se calcularon la media y la desviación estándar de la columna de valores. Luego, se identificaron los valores fuera del rango aceptable:

```python
def detectar_desviacion_estandar(df, columna):
    """Método de desviación estándar"""
    media = df[columna].mean()
    std_dev = df[columna].std()
    outliers = df[(df[columna] > media + 3 * std_dev) | (df[columna] < media - 3 * std_dev)]
    return outliers
```

**Método de Persistencia de Datos**

Este método detecta valores atípicos observando la variación entre valores consecutivos en una serie temporal. Si la diferencia entre dos valores consecutivos es significativamente mayor que la variación típica observada en los datos, el punto se considera atípico.

Se define un umbral basado en:

Media de las diferencias+2×Desviación Estandar de las diferencias

En la función detectar_persistencia, se calcularon las diferencias absolutas entre valores consecutivos usando .diff, y luego se estableció un umbral basado en la media y la desviación estándar de estas diferencias.

```python
def detectar_persistencia(df, columna):
    """Método de persistencia: diferencias grandes entre valores consecutivos"""
    df["Diferencia"] = df[columna].diff().abs()
    umbral = df["Diferencia"].mean() + 2 * df["Diferencia"].std()
    outliers = df[df["Diferencia"] > umbral]
    return outliers.drop(columns=["Diferencia"])
```

**Método de percentiles (1% y 99%)**

Este método define los valores atípicos como aquellos que están en los extremos de la distribución. Se considera atípico cualquier valor menor al percentil 1% o mayor al percentil 99%.

𝑃1=percentil 1 de los datos

𝑃99=percentil 99 de los datos

En la función detectar_percentiles, se calcularon los percentiles 1 y 99 de los datos, y se filtraron los valores fuera de este rango.

```python
def detectar_percentiles(df, columna):
    """Método de percentiles (1% y 99%)"""
    p1 = df[columna].quantile(0.01)
    p99 = df[columna].quantile(0.99)
    outliers = df[(df[columna] < p1) | (df[columna] > p99)]
    return outliers
```

**Método de Puntuación Z (Z-score)**

El Z-score mide cuántas desviaciones estándar un valor se aleja de la media. Se considera atípico un valor si su Z-score es mayor a 3 o menor a -3:

image.png

En la función detectar_z_score, se calculó el Z-score de cada valor y se filtraron aquellos con un Z-score mayor a 3 o menor a -3.

```python

def detectar_z_score(df, columna):
    """Método Z-score"""
    media = df[columna].mean()
    std_dev = df[columna].std()
    df["Z-score"] = (df[columna] - media) / std_dev
    outliers = df[(df["Z-score"] > 3) | (df["Z-score"] < -3)]
    return outliers.drop(columns=["Z-score"])
```

# Aplicando los métodos a cada dataset:

```python

resultados_outliers = {}

for name, df in dataframes.items():
    if df.shape[1] < 2:
        continue
    columna_valor = df.columns[1]
    outliers_iqr = detectar_iqr(df, columna_valor)
    outliers_std = detectar_desviacion_estandar(df, columna_valor)
    outliers_persistencia = detectar_persistencia(df, columna_valor)
    outliers_percentiles = detectar_percentiles(df, columna_valor)
    outliers_zscore = detectar_z_score(df, columna_valor)

    resultados_outliers[name] = {
        "IQR": len(outliers_iqr),
        "Desviación Estándar": len(outliers_std),
        "Persistencia": len(outliers_persistencia),
        "Percentiles": len(outliers_percentiles),
        "Z-score": len(outliers_zscore),
    }


df_resultados = pd.DataFrame(resultados_outliers).T
df_resultados
```
